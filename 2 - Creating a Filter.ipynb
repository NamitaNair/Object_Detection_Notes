## 1. Set Up

### 1. Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

### 2. Visualization Configuration

```python
%matplotlib inline
```

### 3. Data

```python
image = cv2.imread('images/city_hall.jpg')
image_copy = np.copy(image)
```

```python
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
```

## 2. Data Exploration

```python
plt.imshow(image_copy)
```

## 3. Data Preparation

```python
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray, cmap='gray')
```

## 4. Build Vertical and Horizontal Filters

Build a filter that detects vertical edges and ignore horizontal edges

```python
sobel_x = np.array([[-1,0,1],
                   [-2,0,2],
                   [-1,0,1]])
```

```python
sobel_y = np.array([[ -1, -2, -1], 
                   [ 0, 0, 0], 
                   [ 1, 2, 1]])
```

```python
filter_vertical = cv2.filter2D(gray, -1, sobel_x)
filter_horizontal = cv2.filter2D(gray, -1, sobel_y)
```

```python
f, (ax1,ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('Vertical')
ax1.imshow(filter_vertical, cmap='gray')
ax2.set_title('Horizontal')
ax2.imshow(filter_horizontal, cmap='gray')
```

## 5. Create Binary Image

```python
retval, vertical_binary_image = cv2.threshold(filter_vertical, 100, 255, cv2.THRESH_BINARY)
retval, horizontal_binary_image = cv2.threshold(filter_horizontal, 100, 255, cv2.THRESH_BINARY)
```

```python
f, (ax1,ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('Vertical')
ax1.imshow(vertical_binary_image, cmap='gray')
ax2.set_title('Horizontal')
ax2.imshow(horizontal_binary_image, cmap='gray')
```
