## 1. Set Up

### 1. Libraries

```python
import numpy as np

import matplotlib.pyplot as plt

import cv2
```

### 2. Notebook Visualization Configuration

```python
%matplotlib inline
```

### 3. Data

```python
image = cv2.imread('images/brain_MR.jpg')
image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
```

## 2. Data Exploration

```python
plt.imshow(image_copy)
```

## 3. Data Preparation

```python
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)
```

## 4. Apply Low Pass Filter

```python
gray_blur = cv2.GaussianBlur(gray, (5,5), 0)
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('Original')
ax1.imshow(gray, cmap='gray')

ax2.set_title('Blurred')
ax2.imshow(gray_blur, cmap='gray')
```

## 5. Apply High Pass Filter

```python
sobel_x = np.array([[-2, 0, 2],
                  [-1, 0, 1],
                  [-2, 0, 2]])
```

```python
filtered_original = cv2.filter2D(gray, -1, sobel_x)
filtered_blurred = cv2.filter2D(gray_blur, -1, sobel_x)
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('Original')
ax1.imshow(filtered_original, cmap='gray')

ax2.set_title('Blurred')
ax2.imshow(filtered_blurred, cmap='gray')
```
