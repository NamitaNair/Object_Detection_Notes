## 1. Set Up

### 1. Libraries

```python
import numpy as np

import matplotlib.pyplot as plt

import cv2
```

### 2. Notebook Visualization Configuration

```python
%matplotlib inline
```

### 3. Data

```python
image = cv2.imread('images/phone.jpg')
image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
```

## 2. Data Exploration

```python
plt.imshow(image_copy)
```

## 3. Data Preparation

```python
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)
```

## 4. Detect Edges

```python
low_threshold = 50
high_threshold = 100

edges = cv2.Canny(gray, low_threshold, high_threshold)

plt.imshow(edges, cmap='gray')
```

## 5. Find lines using Hough Transforms

```python
rho = 1
theta = np.pi/180
threshold = 60
min_line_length = 100
max_line_gap = 5
```

```python
lines = cv2.HoughLinesP(edges,
                        rho,
                        theta,
                        threshold,
                        np.array([]),
                        min_line_length,
                        max_line_gap)
```

```python
line_image = np.copy(image_copy)
```

```python
for line in lines:
    for x1,y1,x2,y2 in line:
        cv2.line(line_image,(x1,y1),(x2,y2),(255,0,0),5)
```

```python
plt.imshow(line_image)
```
